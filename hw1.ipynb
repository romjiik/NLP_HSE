## 1. Топ 1000 слов по частостности



```python
def lemma_stem_text(word_list):
    lemma = []
    # определяю тэг слова, так как лемматайзеру для корректной работы нужно знать часть речи
    for word, tag in pos_tag(word_list):
      if tag[0].lower() in ['a','n','v']:
        lemma.append(lemmatizer.lemmatize(word, tag[0].lower()))
      else:
        lemma.append(lemmatizer.lemmatize(word))
    return lemma

def review_to_words(raw_review):
    #оставляем только буквы и цифры
    letters_numbers_only = re.sub('[^A-Za-z0-9]', ' ', raw_review)
    #приводим все к нижнему регистру
    # words = np.char.lower(letters_numbers_only.split())
    words = [word.lower() for word in letters_numbers_only.split()]
    #лемматизация
    lemma_words = lemma_stem_text(words)
    return lemma_words
```

**Делаю подсчет на части данных, потому что иначе у меня колаб падает по опертивной памяти:(**

```python
# загружаю список стоп-слов
stopword_list = set(stopwords.words('english'))
# инициализирую лемматайзер
lemmatizer = WordNetLemmatizer()

words = []

for line in raw_corpus[:100000]:
  words.extend(review_to_words(line))

# убираю стоп-слова
words = [word for word in words if word not in stopword_list]
# считаю частотность
count = Counter(words)
print(count.most_common(1000))
```

## 2. Топ 10 имен, имен+фамилий, профессор+имя/фамилия

```python
!pip install svgling
```

```python
names = []
surnames = []
professors = []

# init model
nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)

#add patterns
pattern = [{"LOWER": "professor"}, {"ENT_TYPE": "PERSON"}]
matcher.add("PROFESSOR", [pattern])
pattern_name = [{"ENT_TYPE": "PERSON"}]
matcher.add("PERSON", [pattern_name])
pattern_surname = [{"ENT_TYPE": "PERSON"}, {"ENT_TYPE": "PERSON"}]
matcher.add("SURNAME", [pattern_surname])

clean_corpus = ' '.join(re.findall(r'\b[a-zA-Z]{2,}\b', str(raw_corpus[:10000])))


#check if input is too long
if len(clean_corpus) > 100000:
  nlp.max_length = len(clean_corpus)

# give input to model
reviews = nlp(str(clean_corpus))
# find matches
matches = matcher(reviews)
#iterate through them and add to lists
for match_id, start, end in matches:
    span = Span(reviews, start, end, label=match_id)
    if span.label_ == 'PERSON':
      names.append(span.text)
    elif span.label_ == 'SURNAME':
      surnames.append(span.text)
    elif span.label_ == 'PROFESSOR':
      professors.append(span.text)

# quantity
count_names = Counter(names)
count_surnames = Counter(surnames)
count_professors = Counter(professors)

print(f'Топ-10 имен {count_names.most_common(10)}')
print(f'Топ-10 имен + фамилий {count_surnames.most_common(10)}')
print(f'Топ-10 профессор + имя + фамилия {count_professors.most_common(10)}')
```

```python
from tqdm import tqdm

from tokenizers import Tokenizer, models, trainers
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.trainers import BpeTrainer

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, TensorDataset
from torcheval.metrics.functional import multiclass_f1_score
```

```python
# архитектура модели
class CNN(nn.Module):
    def __init__(self, vocab_size, output_dim, embedding_dim=100, n_filters=100, filter_sizes=[3, 4, 5], dropout=0.5):
        super().__init__()

        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.conv_0 = nn.Conv2d(in_channels=1, out_channels=n_filters, kernel_size=(filter_sizes[0], embedding_dim))
        self.conv_1 = nn.Conv2d(in_channels=1, out_channels=n_filters, kernel_size=(filter_sizes[1], embedding_dim))
        self.conv_2 = nn.Conv2d(in_channels=1, out_channels=n_filters, kernel_size=(filter_sizes[2], embedding_dim))
        self.fc = nn.Linear(len(filter_sizes) * n_filters, output_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        #x = [batch size, sent len]
        embedded = self.embedding(x)

        #embedded = [batch size, sent len, emb dim]
        embedded = embedded.unsqueeze(1)

        #embedded = [batch size, 1, sent len, emb dim]
        conved_0 = F.relu(self.conv_0(embedded).squeeze(3))
        conved_1 = F.relu(self.conv_1(embedded).squeeze(3))
        conved_2 = F.relu(self.conv_2(embedded).squeeze(3))

        #conv_n = [batch size, n_filters, sent len - filter_sizes[n]]
        pooled_0 = F.max_pool1d(conved_0, conved_0.shape[2]).squeeze(2)
        pooled_1 = F.max_pool1d(conved_1, conved_1.shape[2]).squeeze(2)
        pooled_2 = F.max_pool1d(conved_2, conved_2.shape[2]).squeeze(2)

        #pooled_n = [batch size, n_filters]
        cat = self.dropout(torch.cat((pooled_0, pooled_1, pooled_2), dim=1))

        #cat = [batch size, n_filters * len(filter_sizes)]
        return self.fc(cat)
```

```python
# кастомный датасет
class CustomDataset(Dataset):
    def __init__(self, ids, targets):
        self.ids = ids
        self.targets = targets

    def __len__(self):
        return len(self.ids)

    def __getitem__(self, idx):
        id = self.ids[idx]
        target = self.targets[idx]
        return {
            'id': torch.tensor(id),
            'target': torch.tensor(target, dtype=torch.long)
        }
```

```python
# класс для подготовки данных и обучения модели
class Trainer:
    def __init__(self, model, tokenizer, epochs=5, batch_size=64):
        self.device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
        self.model = model.to(self.device)
        self.tokenizer = tokenizer
        self.epochs = epochs
        self.batch_size = batch_size

    def preparation(self, X_train, y_train, X_val, y_val):
        # энкожу каждое предложение
        X_train = tokenizer.encode_batch(X_train)
        X_val = tokenizer.encode_batch(X_val)
        # перевожу инпут в набор айди токенов
        X_train = list(map(lambda x: x.ids, X_train))
        X_val = list(map(lambda x: x.ids, X_val))
        # западим все последовательности до самой длинной последователньости
        max_len_sentence =  max([len(sentence) for sentence in (X_train + X_val)])
        X_train = [sentence + [tokenizer.token_to_id("[PAD]")] * (max_len_sentence - len(sentence)) for sentence in X_train]
        X_val = [sentence + [tokenizer.token_to_id("[PAD]")] * (max_len_sentence - len(sentence)) for sentence in X_val]

        self.train_dataset = CustomDataset(X_train, y_train)
        self.val_dataset = CustomDataset(X_val, y_val)

        self.train_dataloader = DataLoader(self.train_dataset, batch_size=self.batch_size, shuffle=True)
        self.val_dataloader = DataLoader(self.val_dataset, batch_size=self.batch_size, shuffle=False)

        self.optimizer = torch.optim.Adam(self.model.parameters(), lr=1e-3)
        self.loss_fn = torch.nn.CrossEntropyLoss().to(self.device)

    def fit(self):
        self.model = self.model.train()
        losses = []
        # f1_macro_scores = []
        accuracy_scores = []
        correct_predictions = 0

        for step, data in enumerate(self.train_dataloader):
            input_ids = data["id"].to(self.device)
            targets = data["target"].to(self.device)

            outputs = self.model(input_ids)
            preds = torch.argmax(outputs, dim=1)

            loss = self.loss_fn(outputs, targets)

            correct_predictions += torch.sum(preds == targets)
            acc = torch.sum(preds == targets) / len(targets)
            # f1_macro = multiclass_f1_score(preds, targets, num_classes=85, average='macro')

            losses.append(loss.item())
            # f1_macro_scores.append(f1_macro)
            accuracy_scores.append(acc)

            loss.backward()
            self.optimizer.step()
            self.optimizer.zero_grad()


        train_acc = correct_predictions.double() / len(self.train_dataset)
        # train_f1 = torch.mean(torch.tensor(f1_macro_scores, dtype=torch.float32))
        train_loss = torch.mean(torch.tensor(losses, dtype=torch.float32))
        return train_acc, train_loss

    def eval(self):
        self.model = self.model.eval()
        losses = []
        # f1_macro_scores = []
        correct_predictions = 0

        with torch.no_grad():
            for data in self.val_dataloader:
                input_ids = data["id"].to(self.device)
                targets = data["target"].to(self.device)

                outputs = self.model(input_ids)

                preds = torch.argmax(outputs, dim=1)
                loss = self.loss_fn(outputs, targets)
                # f1_macro = multiclass_f1_score(preds, targets, num_classes=85, average='macro')
                correct_predictions += torch.sum(preds == targets)
                losses.append(loss.item())
                # f1_macro_scores.append(f1_macro)

        val_acc = correct_predictions.double() / len(self.val_dataset)
        # test_f1 = torch.mean(torch.tensor(f1_macro_scores, dtype=torch.float32))
        val_loss = torch.mean(torch.tensor(losses, dtype=torch.float32))
        return val_acc, val_loss

    def train(self):
        best_acc = 0
        for epoch in tqdm(range(self.epochs)):
            print('\n' + f'Epoch {epoch + 1}/{self.epochs}')
            train_acc, train_loss = self.fit()
            print(f'Train loss {train_loss}, Train accuracy {train_acc}')

            val_acc, val_loss = self.eval()
            print('\n' + f'Val loss {val_loss}, Val accuracy {val_acc}')
            print('-' * 10)

            if val_acc > best_acc:
                best_acc = val_acc
            # print(f'Epoch: {epoch+1}, Train Loss: {train_loss:.3f}, Train Acc: {train_acc*100:.2f}%, Train F1: {train_f1*100:.2f}%')
            # print(f'Val. Loss: {val_loss:.3f}, Val. Acc: {val_acc*100:.2f}%, Val. F1: {val_f1*100:.2f}%')
        print('Модель обучена')
        print('Accuracy =', float(best_acc))

    def predict(self, X_test):
        X_test = tokenizer.encode_batch(X_test)
        # перевожу инпут в набор айди токенов
        X_test = list(map(lambda x: x.ids, X_test))
        X_test = torch.tensor(X_test).to(self.device)
        outputs = self.model(X_test)
        prediction = outputs.cpu().detach()
        return prediction
```

```python
# инициализирую токенизатор
tokenizer = Tokenizer(models.BPE(unk_token="[UNK]"))
tokenizer.pre_tokenizer = Whitespace()
trainer = BpeTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])
tokenizer.train_from_iterator(df_training['review'], trainer=trainer)
# инициализирую модель
vocab_size = len(tokenizer.get_vocab())
output_dim = df_training['charm'].nunique()

cnn_model = CNN(vocab_size, output_dim)

trainer = Trainer(cnn_model, tokenizer)
```

# Обучаю RNN

```python
class RNN(nn.Module):
    def __init__(self, vocab_size, output_dim, hidden_dim=64, embedding_dim=100, n_layers=1, dropout=0):
        super(RNN, self).__init__()

        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, num_layers=n_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = self.embedding(x)
        output, hidden = self.rnn(x)
        x = self.fc(output[:, -1, :])
        return x
```

```python
rnn_model = RNN(vocab_size, output_dim)

trainer_rnn = Trainer(rnn_model, tokenizer)
```
